```python
import pandas as pd
from pytrends.request import TrendReq

pytrend = TrendReq(tz=360) # US central time
```

```python
internal_codes = pd.read_csv("internal_codes.csv")
```

```python
mid_codes_dict = {"topic":[], "mid_code":[]}
```

```python
codes = ['covid', 'covid vaccine', 'covid quarantine mental health', 'covid housing', 'covid economy', 'covid mask guidelines', 'social distancing', 'social anxiety disorder']
idx = 7      # which item in codes list I want to look at
pytrend.build_payload([codes[idx]], geo='US', timeframe = '2020-01-07 2022-01-07')
```

```python
related_topic = pytrend.related_topics()
top = list(related_topic.values())[0]['top']
dftop = pd.DataFrame(top)
dftop.head(20)
```

```python
idx = 2    # index in dftop that we want code from
mid_codes_dict['topic'].append(dftop.topic_title[idx])
mid_codes_dict['mid_code'].append(dftop.topic_mid[idx])
mid_codes_dict
```

```python
# for generating dataframe for the first time
internal_codes = pd.DataFrame(mid_codes_dict)
internal_codes
```

```python
# for adding to the dataframe
idx = 2
mid_codes_dict = {}
mid_codes_dict['topic'] = dftop.topic_title[idx]
mid_codes_dict['mid_code'] = dftop.topic_mid[idx]
mid_codes_dict
internal_codes = internal_codes.append(mid_codes_dict, ignore_index=True)
internal_codes
```

```python
internal_codes.to_csv('internal_codes.csv', index=False)
```
